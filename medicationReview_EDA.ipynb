<p><b>5.2 Cleaning the text</b></p>

<p>Since the reviews columns have been cleaned, we will focus on the text present in the columns in the reviews column by cleaning the content by removing all special characters, non-ASCII characters, trailing whitespaces, and replacing multiple spaces with a single space and two dots into one dot.</p>

```python
def review_clean(review): 
    # changing to lower case
    lower = review.str.lower()
    
    # Removing all the special characters
    special_remove = lower.str.replace(r'[^\w\d\s]',' ')
    
    # Removing all the non ASCII characters
    ascii_remove = special_remove.str.replace(r'[^\x00-\x7F]+',' ')
    
    # Removing leading and trailing whitespaces
    whitespace_remove = ascii_remove.str.replace(r'^\s+|\s+?$','')
    
    # Replacing multiple spaces with single Space
    multiw_remove = whitespace_remove.str.replace(r'\s+',' ')
    
    # Replacing two or more dots with one
    dataframe = multiw_remove.str.replace(r'\.{2,}', ' ')
    
    return dataframe
```

<p>The function review_clean is first converts thje text into lower and follows the removing and replacing through the result and it follows a continuous step-by-step process as the output for one is used as the input to return the data frame.</p>

```python
#passing the reviews of dep_data throught the review_clean function
dep_data['reviews'] = review_clean(dep_data['reviews'])
```

```python
#printing the first five rows to show th implementation of review_clean function on reviews columns
dep_data.head(5)
```

<p>On reading through the text in the reviews column we can see the words such as i, but,my, and many more. These stop words should be removed as some analyses will look at the frequency of words and without removing stop words, the repetition of words such as  ”I” will appear as the highest frequency despite not being meaningful for analysis.</p> 

```python
#removing stops words using lambda using split to divide the sentences into words 
stop_words = stopwords.words('english')
dep_data['reviews'] = dep_data['reviews'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop_words)]))
```

<p>Since we have removed the stop words, we will try to get the frequency of unique words present in the texts of the review column. This implementation is done using the counter function which counts the unique words.</p>

```python
#get frequnecy of unique words
unique_words = Counter()
dep_data['reviews'].str.lower().str.split().apply(unique_words.update)
```

```python
# printing the first 20 text of review column 
dep_data['reviews'][0:20]
```

<p><b>5.3 Lemmantization</b></p>

<p>We can still distill the words to make them more meaning for the analyis either through lemmatization or stemming. Lemmantization return a root word while stemming removes suffix to create a stem. 

While in later analysis through visualization, it is preferred to able to see meaningful whole words rather than unfinished words to make the visualization more impactful. Hence it is better to implemn lemmatization to get the root words.</p>

```python
#lemmatizer
lemmatizer=WordNetLemmatizer()

# lemmatizer function
def lemmatize_text(review):
    lemmatized_tokens = [lemmatizer.lemmatize(word) for word in review]
    return ' '.join(lemmatized_tokens)
```

<p> I have created an instance of lemmantizer and lemmantize_text function. </p>

```python
#creating a row named review_clean and calling the lemmatize_text function
dep_data['review_clean'] = dep_data['reviews'].apply(lambda x: " ".join(lemmatizer.lemmatize(word) for word in x.split()))
```

<p> I have create a new column named as review_clean storing the lemmatized text. Creating a new column, review_clean prevents overlapping on the reviews column as creating the root of word might end with unexpected results.</p>

```python
#printing the top 20 rows of review_clean
dep_data['review_clean'][0:20]
```
